como hay columnas dupicadas:

eliminamos las columnas duplicadas:

```python
df_sin_duplicados = df.loc[:, ~df.columns.duplicated()]
```

```python
df_sin_duplicados.info()
```

```python
data=pd.DataFrame(df_sin_duplicados)
```

Buscamos locales de la franquicia mc donalds en la columna name :

```python
resultados = data[data['name'].str.contains('McDonald', case=False)]
resultados.head()
```

```python
resultados = resultados.assign(categories = resultados.categories.str.split(', ')).explode('categories')
resultados.head()
```

Guardamos en un archivo csv los datos de las franquicias mc donalds:

```python
resultados.to_csv('bussines_yelp.csv',index=False)
```

**REVIEW.JSON**

abrimos el archivo json en partes con el siguiente codigo :

```python
chunk_size = 5000  # Adjust this based on your memory availability
json_reader = pd.read_json('review.json', lines=True, chunksize=chunk_size)
df_list = []
for chunk in json_reader:
    df_list.append(chunk)
df_rev = pd.concat(df_list, ignore_index=True)
```

```python
df_rev.head()
```

```python
df_rev.info()
```

eliminamos duplicados de las columnas 'review_id' y 'text':

```python
df_rev=df_rev.drop_duplicates(subset=['review_id','text'],keep="first")
df_rev['id_review']=df_rev.index
```

```python
df_rev.drop(['useful','funny','cool'],axis=1,inplace=True)
```

```python
df_rev.head()
```

Divide la columna date en hora y fecha 

```python
df_rev['hour'] = pd.to_datetime(df_rev['date']).dt.strftime('%H:%M:%S')

df_rev['date'] = pd.to_datetime(df_rev['date']).dt.strftime('%Y-%m-%d')
```

```python
df_rev.head()
```

```python
for chunk in df_rev:
    year_to_filter = '2016'
    filtered_df = df_rev[df_rev['date'].str.startswith(year_to_filter)]
    chunk_filter = filtered_df[filtered_df['date'] > year_to_filter] 
    chunk_filter=chunk_filter.drop(['review_id','id_review'],axis=1)
    chunk_filter.to_parquet('review_yelp.parquet',  compression="snappy")
```

```python
yelp = resultados.merge(chunk_filter, on='business_id', how='inner')
yelp.head()
```

```python
yelp.to_csv('BUSINESS_YELP_YORK.csv')
```
