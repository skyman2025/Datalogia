ANALIZAMOS CON EL MODELO DE ANALISIS DE SENTIMIENTOS  NLTK ( «Natural Language Toolkit») 
sobre la columna 'text' de review

```python
analizador = SentimentIntensityAnalyzer()
```

```python
reseña=review['text']
```

```python
sentences = []
for text in reseña:
        sentence_list = sent_tokenize(text)
        sentences.extend(sentence_list)
```

CONSIDERACIONES A TOMAR EN CUENTA AL MOMENTO DE DEVOLVER LOS VALORES :

Para cada frase se obtienen varias puntuaciones diferentes, que podremos ver en la salida un poco más abajo. Pero antes resumimos los diversos criterios para las puntuaciones.

neg (negativa): es un valor entre cero y uno, para decirnos lo negativa que sería esta frase.

neu (neutral): este segundo valor nos indica la neutralidad de una frase, también en una puntuación entre cero y uno.Locual no seria de mucho interes para el analisis

pos (positiva): Igualmente que los anteriores, pero indicando lo positiva que encuentra una frase.

compound: este es un valor entre -1 y 1 que viene a indicar de una única vez si la frase es positiva o negativa. Valores próximos a -1 indican que es muy negativa, próximos a cero indicarían que es neutra y próximos a 1 sería muy positiva

```python
lista=[]
for indice,sentence in enumerate(sentences):
    scores = analizador.polarity_scores(sentence)
    #scores={'neg': 0.0, 'neu': 1.0, 'pos': 0.0, 'compound': 0.0}
    for key, value in scores.items():
      if key == 'compound':
        if scores[key]> 0.6:
          lista.extend('5')
        elif 0.6 <= scores[key] < 0.2:
          lista.extend('4')
        elif -0.2 < scores[key] <= 0.2:
          lista.extend('3')
        elif -0.6 <= scores[key] < -0.2:
          lista.extend('2')
        elif scores[key] < -0.6:
          lista.extend('1')
```

```python
review['nltk_stars']=pd.DataFrame(lista)
```

```python
review['nltk_stars']=(review['nltk_stars']).astype('float')
```

```python
review.drop(columns=['text'],axis=1,inplace=True)
```

por cada bussines_id se hace un promedio de sus respectivas puntuaciones en la columna 'stars':

```python
review_stars=review.groupby('business_id').mean()
```

```python
review_stars.reset_index(inplace=True)
```

RANDOM FOREST

Se separan las variables de pendientes 'y' de las independientes 'X':

```python
X = stars.drop('promedio', axis=1)
y = stars['promedio']
```

Eleccion de parametros y entrenamiento:

```python
model_forest=RandomForestRegressor(n_estimators=305,max_depth=60,min_samples_split=20,max_features=37)
```

Se separa la data en entrenamiento y tests con una proporcion de 70 entrenamiento y 30 a testeo:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
model_forest.fit(X_train, y_train)
```

Pruebas:

```python
y_pred = model_forest.predict(X_test)
```

Este hiperparametro devuelve un  número entre 0 y 1 que indica cuán bien se ajusta el modelo de bosque aleatorio a los datos de prueba proporcionados

```python
model_forest.score(X_test, y_test)
```

```python
pruebaforest=y_test.reset_index()
pruebaforest['star_pred']=y_pred
pruebaforest
```

```python
from sklearn.metrics import mean_squared_error
```

Hiperparametro de error cuadratico:

```python
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio:", mse)
```
